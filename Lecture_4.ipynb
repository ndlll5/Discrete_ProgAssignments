```python
import matplotlib.pyplot as plt
import networkx as nx

seed = 55
# Use seed for reproducibility
G = nx.Graph()
plt.figure(figsize=(10, 8))
# Add nodes
G.add_nodes_from(["A", "B", "C", "D", "E", "F", "G"])
# Add self-loops to the remaining nodes
edgelist = [("A", "C"), ("A", "I"), ("C", "B"), ("C", "E"), ("C", "D"), ("B", "E"), ("I", "E"), ("E", "F"),
            ("F", "G"), ("G", "E")]
G.add_edges_from(edgelist)
pos = nx.planar_layout(G)  # Seed for reproducible layout

# Highlighted edges
highlighted_edges = [("I", "A"), ("A", "C"), ("C", "B"), ("C", "E"), ("C", "D"), ("E", "F")]

# Draw network graph
nx.draw(G, pos=pos, with_labels=True, node_size=500, node_color='pink', alpha=0.9)

# Draw all edges
nx.draw_networkx_edges(G, pos, edgelist=G.edges(), edge_color='black', width=1)

# Draw highlighted edges with different color
nx.draw_networkx_edges(G, pos, edgelist=highlighted_edges, edge_color='green', width=5)

# Draw edge labels
edge_labels = {("A", "C"): 10, ("A", "I"): 5, ("C", "B"): 20, ("C", "E"): 15, ("C", "D"): 30, ("B", "E"): 25,
               ("I", "E"): 40, ("E", "F"): 35, ("F", "G"): 50, ("G", "E"): 45}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')

plt.show()
```
